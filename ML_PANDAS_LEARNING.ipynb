# merge

```python
df1=pd.DataFrame({'Roll No.':[1,2,3,4,5],'Physics':[34,67,34,89,12]})
df1
```

```python
df2=pd.DataFrame({'Roll No.':[1,2,3,4,5],'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1,df2,on='Roll No.')
```

```python
pd.merge(df1,df2) #on by default takes intersecting column
```

```python
df3=pd.DataFrame({'Roll No.':[1,2,3,6,7],'Physics':[34,67,34,89,12]})
df4=pd.DataFrame({'Roll No.':[1,2,3,4,5],'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3,df4)#takes rollnumbers which are intersecting
```

```python
pd.merge(df3,df4,how='left')#combines with left table values
```

```python
pd.merge(df3,df4,how='right')#combines with right table values
```

```python
pd.merge(df3,df4,how='outer')#all combinations
```

# Pivot table

```python
df=pd.read_csv(r'/Users/rohanlalgudivada/Desktop/sample2.csv')
df.head(5)
```

```python
pd.pivot_table(df,index='Branch')#average values of the respective index
```

```python
pd.pivot_table(df,index='Branch',aggfunc='sum')#sum of all values
```

```python
pd.pivot_table(df,index='Branch',aggfunc='count')#count of respective column
```

```python
pd.pivot_table(df,index='Branch',aggfunc='max')#maximum in each 
```
