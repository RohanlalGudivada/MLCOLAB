# Bar Plot

```python
plt.bar(sub,marks,color='g')
plt.show()
```

# Histogram

```python
m50=np.random.randint(0,100,(50))
m50
```

```python
plt.hist(m50)
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(m50,bins=bins)
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(m50,bins=bins,color='y')
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(m50,bins=bins,color='y',orientation='horizontal')
plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(m50,bins=bins,color='y',rwidth=0.4)
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(m50,bins=bins,color='y',histtype='step')#types-bar(default),step,barstacked,stepfilled
plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
m501=np.random.randint(0,100,(50))
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist([m50,m501],bins=bins,color=['orange','b'])
plt.xticks(np.arange(0,100,5))
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks of 2 classes')
plt.show()
```

# Pie Charts

```python
classes=['PHY','CHE','MATHS','SCIENCE','SST']
marks=[89,45,78,23,90]
```

```python
plt.pie(marks,labels=classes)#in pie charts colours to be given as colors
plt.show()
```

```python
plt.pie(marks,labels=classes,autopct='%0.2f%%')#autopct displays the percentage 0.2 tells the no.of decimal places
plt.show()
```

```python
exp1=[0.1,0.2,0,0,0]#explode is for making the graph piece divided
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1)
plt.show()
```

```python
exp1=[0.1,0.2,0,0,0]
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1,shadow=True)#shadow can be seen
plt.show()
```

```python
exp1=[0.1,0.2,0,0,0]
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1,shadow=True,radius=1.6)
plt.show()#radius makes the size of circle
```

```python
textprops={'fontsize':10,'color':'k'}#changes the size of written percentage in the graph
exp1=[0.1,0.2,0,0,0]
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1,textprops=textprops)
plt.show()
```

```python
textprops={'fontsize':10,'color':'k'}
wedgeprops={'linewidth':3,'linestyle':'--','edgecolor':'white'}#wedgeprops decides properties of the wedge
exp1=[0.1,0.2,0,0,0]
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1,textprops=textprops,wedgeprops=wedgeprops)
plt.show()
```

```python
plt.figure(figsize=(6,6))
textprops={'fontsize':10,'color':'k'}
wedgeprops={'linewidth':3,'linestyle':'--','edgecolor':'white'}
exp1=[0.1,0.2,0,0,0]
plt.pie(marks,labels=classes,autopct='%0.2f%%',explode=exp1,textprops=textprops,wedgeprops=wedgeprops)
plt.legend()#legend describes the graph the table seen
plt.show()
```
